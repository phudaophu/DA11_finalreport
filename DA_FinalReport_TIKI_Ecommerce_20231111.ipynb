     product_id: 	    same with book_data file
     comment_id: 	    each comment has individual id
     title: 	        keywork of comment
     thank_count: 	    number of like of other people
     customer_id: 	    each customer has individual id
     rating:     	    average rating of the comment
     content:    	    same with it's name

### 1. Sản phẩm sách có số lượt bán nhiều nhất và doanh thu tương ứng

```python
w_Output1 = widgets.Output(layout={'border': '1px solid black', 'height':"1100px", 'width': '990px'})
```

```python
def ve_sanpham_quantity(N=10):
    df_products_quantity = df_general[['title','quantity','total_sales']].sort_values(by='quantity',ascending=False)[:N]
    df_others_quantity = df_general[['title','quantity','total_sales']].sort_values(by='quantity',ascending=False)[N:]
    sum_others_quantity = df_others_quantity['quantity'].sum()
    sum_others_sales = df_others_quantity['total_sales'].sum()
    df_others_quantity = pd.DataFrame({'title':['Others'],'quantity':[sum_others_quantity],'total_sales':[sum_others_sales]})
    df_products_quantity = pd.concat([df_products_quantity,df_others_quantity],axis=0)
    
    
    with w_Output1:
        w_Output1.clear_output()
        fig01 = plt.figure(figsize=(10,15))
        ax01 = fig01.add_subplot(2,1,1)
        ax01.barh(y=df_products_quantity['title'],width=df_products_quantity['quantity'], color='green', alpha=0.7,height=0.7)
        ax01.set_title(f"Biều đồ thể hiện top {N} sản phẩm best-sellers")

        ax02 = fig01.add_subplot(2,1,2)
        ax02.barh(y=df_products_quantity['title'],width=df_products_quantity['total_sales'], color='pink', alpha=0.7, height=0.7)
        ax02.set_title("Biều đồ thể hiện doanh thu tương ứng của các sản phẩm best-sellers")
        plt.show()
```

### 2. Tác giả,nhà xuất bản, thể loại có doanh thu cao nhất 

```python
w_Output2 = widgets.Output(layout={'border': '1px solid black', 'height':"1400px", 'width': '950px'})
```

```python
def ve_obj_total_sales(N=10,choose='authors',show_details=False):
    df_obj_quantity = df_general.groupby(by=choose)['total_sales'].sum().reset_index()
    df1 = df_obj_quantity.sort_values(by='total_sales',ascending=False)[:N]

    df1_others = pd.DataFrame({choose:'Others','total_sales':[df_obj_quantity['total_sales'][N:].sum()]})
    df1=pd.concat([df1,df1_others])
    df1.reset_index()

    df_obj_cntbook = df_general.groupby(by=choose)['title'].count().reset_index()
    sum_title = df_obj_cntbook['title'].sum()
    df2 = df1.merge(df_obj_cntbook,on=choose)
    df2_others = pd.DataFrame({choose:'Others','total_sales':[df_obj_quantity['total_sales'][N:].sum()], 'title':[sum_title -df2['title'].sum()]}) 

    df2 = pd.concat([df2,df2_others])
    df2.rename(columns={'title':'book_number'},inplace=True)
    df2.reset_index(inplace=True,drop=True)
    with w_Output2:
        w_Output2.clear_output()
        fig01 = plt.figure(figsize=(8,5))
        ax01 = fig01.add_subplot(1,2,1)
        ax01.barh(y=df2[choose],width=df2['total_sales'], color='green', alpha=0.7)
        ax01.set_title(f"Biều đồ thể hiện top {N} {choose} có doanh số sách cao nhất")
        
        ax02 = fig01.add_subplot(1,2,2)
        ax02.pie(df2['total_sales'],pctdistance=1.1,autopct='%1.0f',textprops={'fontsize':7})
        ax02.legend(labels=df2[choose], loc=(1,0.5))
        
        plt.show()
        if(show_details):
            display(df2)
    
    
```

```python
w_Slider2 = widgets.IntSlider(value=10, min=10, max=30, step=1, continuous_update = False)
w_Select2 = widgets.Dropdown(options=['authors','manufacturer','category'],value='authors')
w_CheckBox2 = widgets.Checkbox(value=False, description='Show details',disabled=False, indent=False )
w_Box2 = widgets.HBox([w_Slider2,w_Select2,w_CheckBox2])
```

```python
ve_obj_total_sales = widgets.interactive(ve_obj_total_sales,N=w_Slider2, choose = w_Select2,show_details=w_CheckBox2)
```

```python
df_X = df_XY[['discount','n_review','pages','rating4','rating5','avg_rating']]
df_Y = df_XY['quantity']
```

```python
x = df_X.values.tolist()
y = df_Y.values.tolist()
```

```python
y[0]
```

#### Chia dữ liệu thành 2 tập: traing và test

```python
np.random.seed(52)
rand_idx = np.arange(len(x))
np.random.shuffle(rand_idx)

print(rand_idx)
print(len(rand_idx))
```

```python
N_TRAIN = round(0.8*len(rand_idx))
print(N_TRAIN)
train_indices = rand_idx[:N_TRAIN]
test_indices  = rand_idx[N_TRAIN:]
```

```python
x_train = [x[i] for i in train_indices]
y_train = [y[i] for i in train_indices]
print(x_train[:5])
print(y_train[:5])
print(len(x_train))
print(len(y_train))
```

```python
x_test = [x[i] for i in test_indices]
y_test = [y[i] for i in test_indices]
print(x_test[:5])
print(y_test[:5])
print(len(x_test))
print(len(y_test))
```

#### Khởi tạo và huấn luyện mô hình

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
# Thực hiện huấn luyện mô hình
model.fit(x_train, y_train)
```

```python
# Kiểm tra các hệ số của mô hình
print(model.coef_)
print(model.intercept_)
```

```python
y_hat = model.predict(x_test)
print(len(y_hat))
```

```python
plt.figure(figsize=[20,7]) 
plt.scatter(y_hat,y_test)
plt.show()
```

```python
r2 = model.score(x_train,y_train)
q2 = model.score(x_test,y_test)
print(r2)
print(q2)
```

#### Hệ số r2 và q2 chênh lệch rất lớn

```python
# Tính lỗi trung bình residual 
error_mean = np.sqrt((y_hat - y_test)**2).mean()
print(error_mean)
```
